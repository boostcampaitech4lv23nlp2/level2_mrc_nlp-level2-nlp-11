## TODO!
이 데이터셋을 우리가 만든 데이터셋과 동일한 포맷으로 만들어야 한다!

그러기 위해서는 paragraph를 언패킹해야 한다.

```python
book_df = book_df.explode('paragraphs')
```

```python
book_df = book_df.reset_index()
```

```python
book_df
```

```python
book_df[['context', 'qas']] = pd.json_normalize(book_df['paragraphs'])
book_df
```

```python
book_df = book_df.explode('qas').reset_index()
book_df
```

```python
book_df[['question', 'id', 'is_impossible', 'answers']] = pd.json_normalize(book_df['qas'])
book_df
```

```python
book_df = book_df.drop(['index', 'time', 'agency', 'year', 'KDC', 'paragraphs', 'qas', 'level_0'], axis=1)
book_df
```

```python
book_df = book_df[book_df.is_impossible == False].reset_index().drop(['is_impossible', 'index'], axis=1)
book_df
```

모든 answer가 단 하나씩만 존재한다.

```python
book_df = book_df.rename(columns={'content_id':'document_id'}).explode('answers')
```

```python
book_df[['text', 'answer_start']] = pd.json_normalize(book_df['answers'])
```

```python
book_df
```

```python
book_df = book_df.drop(['answers'], axis=1)
book_df['answers'] = book_df.apply(lambda x: {'text': [x['text']], 'answer_start':[x['answer_start']]}, axis=1)
book_df
```

```python
book_df.to_csv('../data/book/dataset.csv')
```
